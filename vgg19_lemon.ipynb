```python
# ============================================================
# STEP 1: CLEAR MEMORY
# ============================================================
import gc
import tensorflow as tf
tf.keras.backend.clear_session()
gc.collect()
print("✅ Memory cleared!")

# ============================================================
# STEP 2: IMPORTS
# ============================================================
from tensorflow.keras.applications import VGG19
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import (EarlyStopping,
                                        ReduceLROnPlateau,
                                        ModelCheckpoint)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score,
                             classification_report, confusion_matrix)
import seaborn as sns
import os
import shutil

print("TensorFlow:", tf.__version__)
print("GPU:", tf.config.list_physical_devices('GPU'))

# ============================================================
# STEP 3: PATHS
# ============================================================
ORIGINAL_PATH = "/content/drive/MyDrive/lemon/Original Dataset"

print("\nClasses:", sorted(os.listdir(ORIGINAL_PATH)))

print("\n--- Image counts ---")
total_images = 0
for cls in sorted(os.listdir(ORIGINAL_PATH)):
    count = len(os.listdir(os.path.join(ORIGINAL_PATH, cls)))
    total_images += count
    print(f"  {cls}: {count} images")
print(f"\nTotal: {total_images} images")

# ============================================================
# STEP 4: SETTINGS
# ============================================================
IMG_SIZE    = (224, 224)
BATCH_SIZE  = 16        # Keep small — VGG19 is heavy
SEED        = 42
NUM_CLASSES = 9

# ============================================================
# STEP 5: COPY TO LOCAL DISK
# ============================================================
LOCAL_ORIG = "/content/Original_dataset"

if not os.path.exists(LOCAL_ORIG):
    print("\nCopying dataset to local disk...")
    shutil.copytree(ORIGINAL_PATH, LOCAL_ORIG)
    print("✅ Dataset copied!")
else:
    print("✅ Already on local disk!")

# ============================================================
# STEP 6: LOAD & SPLIT — 70% Train | 15% Val | 15% Test
# ============================================================
print("\nLoading Original dataset...")

train_ds = tf.keras.utils.image_dataset_from_directory(
    LOCAL_ORIG,
    validation_split=0.30,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_test_ds = tf.keras.utils.image_dataset_from_directory(
    LOCAL_ORIG,
    validation_split=0.30,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# Split val_test 50/50
total_val_test = tf.data.experimental.cardinality(val_test_ds).numpy()
val_size       = total_val_test // 2

val_ds  = val_test_ds.take(val_size)
test_ds = val_test_ds.skip(val_size)

class_names = train_ds.class_names
print(f"\nClasses ({len(class_names)}):", class_names)

train_batches = tf.data.experimental.cardinality(train_ds).numpy()
val_batches   = tf.data.experimental.cardinality(val_ds).numpy()
test_batches  = tf.data.experimental.cardinality(test_ds).numpy()

print(f"\nTrain batches : {train_batches} (~{train_batches * BATCH_SIZE} images)")
print(f"Val batches   : {val_batches}   (~{val_batches   * BATCH_SIZE} images)")
print(f"Test batches  : {test_batches}  (~{test_batches  * BATCH_SIZE} images)")

# ============================================================
# STEP 7: AUGMENTATION
# ============================================================
augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.3),
    tf.keras.layers.RandomZoom(0.2),
    tf.keras.layers.RandomBrightness(0.2),
    tf.keras.layers.RandomContrast(0.2),
], name="augmentation")

train_ds = train_ds.map(
    lambda x, y: (augmentation(x, training=True), y),
    num_parallel_calls=tf.data.AUTOTUNE
)

# ============================================================
# STEP 8: OPTIMIZATION
# ============================================================
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.shuffle(500).prefetch(buffer_size=AUTOTUNE)
val_ds   = val_ds.prefetch(buffer_size=AUTOTUNE)
test_ds  = test_ds.prefetch(buffer_size=AUTOTUNE)

# ============================================================
# STEP 9: BUILD VGG19
# ============================================================
base_model = VGG19(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)
base_model.trainable = False  # Frozen

print(f"\nVGG19 loaded!")
print(f"Total layers: {len(base_model.layers)}")

# Build model
inputs = tf.keras.Input(shape=(224, 224, 3))

# VGG19 preprocessing
x = tf.keras.applications.vgg19.preprocess_input(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)

model_vgg19 = models.Model(inputs, outputs)
model_vgg19.summary()

# Parameter summary
total     = model_vgg19.count_params()
trainable = sum([tf.size(w).numpy()
                 for w in model_vgg19.trainable_weights])
frozen    = total - trainable
print(f"\nTotal params    : {total:,}")
print(f"Trainable params: {trainable:,}  ← Your head")
print(f"Frozen params   : {frozen:,}  ← VGG19 base")
print(f"% Frozen        : {frozen/total*100:.1f}%")

# ============================================================
# STEP 10: COMPILE
# ============================================================
model_vgg19.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# ============================================================
# STEP 11: CALLBACKS
# ============================================================
checkpoint_path = '/content/drive/MyDrive/lemon/best_vgg19_original.keras'

callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=7,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        patience=3,
        factor=0.2,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        filepath=checkpoint_path,
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]

# ============================================================
# STEP 12: TRAIN
# ============================================================
print("\n" + "="*50)
print("TRAINING VGG19 (Original Dataset Only)...")
print("Split: 70% Train | 15% Val | 15% Test")
print("="*50)

history = model_vgg19.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50,
    callbacks=callbacks
)

# ============================================================
# STEP 13: PLOT HISTORY
# ============================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(history.history['accuracy'],     'b-', label='Train Accuracy')
ax1.plot(history.history['val_accuracy'], 'r-', label='Val Accuracy')
ax1.set_title('VGG19 (Original) — Accuracy', fontweight='bold')
ax1.set_xlabel('Epoch'); ax1.set_ylabel('Accuracy')
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(history.history['loss'],     'b-', label='Train Loss')
ax2.plot(history.history['val_loss'], 'r-', label='Val Loss')
ax2.set_title('VGG19 (Original) — Loss', fontweight='bold')
ax2.set_xlabel('Epoch'); ax2.set_ylabel('Loss')
ax2.legend(); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/lemon/vgg19_original_history.png', dpi=150)
plt.show()

# ============================================================
# STEP 14: EVALUATE ON TEST SET
# ============================================================
print("\n" + "="*50)
print("EVALUATING ON TEST SET (15% Original)...")
print("="*50)

y_true, y_pred = [], []
for images, labels in test_ds:
    preds = model_vgg19.predict(images, verbose=0)
    y_pred.extend(np.argmax(preds, axis=1))
    y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

acc  = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
rec  = recall_score(y_true, y_pred, average='macro', zero_division=0)
f1   = f1_score(y_true, y_pred, average='macro', zero_division=0)

print(f"\n{'='*40}")
print(f"  VGG19 — Original Dataset")
print(f"{'='*40}")
print(f"  Accuracy : {acc:.4f} ({acc*100:.2f}%)")
print(f"  Precision: {prec:.4f} ({prec*100:.2f}%)")
print(f"  Recall   : {rec:.4f} ({rec*100:.2f}%)")
print(f"  F1-Score : {f1:.4f} ({f1*100:.2f}%)")
print(f"{'='*40}")

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# ============================================================
# STEP 15: CONFUSION MATRIX
# ============================================================
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds',
            xticklabels=class_names,
            yticklabels=class_names)
plt.title('VGG19 (Original) — Confusion Matrix',
          fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/lemon/vgg19_original_cm.png', dpi=150)
plt.show()

# ============================================================
# STEP 16: SAVE MODEL & RESULTS
# ============================================================
model_vgg19.save('/content/drive/MyDrive/lemon/vgg19_original_final.keras')

results_text = f"""
VGG19 — Original Dataset Only
========================================
Dataset     : Original (1,354 images)
Split       : 70% Train | 15% Val | 15% Test
Augmentation: Runtime augmentation applied
========================================
Accuracy : {acc:.4f} ({acc*100:.2f}%)
Precision: {prec:.4f} ({prec*100:.2f}%)
Recall   : {rec:.4f} ({rec*100:.2f}%)
F1-Score : {f1:.4f} ({f1*100:.2f}%)
========================================
"""

with open('/content/drive/MyDrive/lemon/vgg19_original_results.txt', 'w') as f:
    f.write(results_text)

print(results_text)

# ============================================================
# STEP 17: FULL COMPARISON TABLE
# ============================================================
print("\n" + "="*55)
print("  RESULTS — Original Dataset Experiments")
print("="*55)
print(f"  {'Model':<20} {'Accuracy':>10} {'F1-Score':>10}")
print("-"*55)
print(f"  {'EfficientNetB3':<20} {'?%':>10} {'?%':>10}")
print(f"  {'DenseNet121':<20} {'?%':>10} {'?%':>10}")
print(f"  {'VGG19':<20} {acc*100:>9.2f}% {f1*100:>9.2f}%")
print("="*55)

# Clear memory
del model_vgg19
tf.keras.backend.clear_session()
gc.collect()
print("\n✅ Memory cleared!")
print("✅ All VGG19 outputs saved to Drive!")
```

```python
# ============================================================
# STEP 1: CLEAR MEMORY
# ============================================================
import gc
import tensorflow as tf
tf.keras.backend.clear_session()
gc.collect()
print("✅ Memory cleared!")

# ============================================================
# STEP 2: IMPORTS
# ============================================================
from tensorflow.keras.applications import VGG19
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import (EarlyStopping,
                                        ReduceLROnPlateau,
                                        ModelCheckpoint)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score,
                             classification_report, confusion_matrix)
import seaborn as sns
import os
import shutil

print("TensorFlow:", tf.__version__)
print("GPU:", tf.config.list_physical_devices('GPU'))

# ============================================================
# STEP 3: PATHS
# ============================================================
AUGMENTED_PATH = "/content/drive/MyDrive/lemon/Augmented Dataset"
ORIGINAL_PATH  = "/content/drive/MyDrive/lemon/Original Dataset"

print("\nClasses:", sorted(os.listdir(ORIGINAL_PATH)))

print("\n--- Image counts (Original Test Set) ---")
total = 0
for cls in sorted(os.listdir(ORIGINAL_PATH)):
    count = len(os.listdir(os.path.join(ORIGINAL_PATH, cls)))
    total += count
    print(f"  {cls}: {count} images")
print(f"Total: {total} images")

# ============================================================
# STEP 4: SETTINGS
# ============================================================
IMG_SIZE    = (224, 224)
BATCH_SIZE  = 16        # Keep small — VGG19 is heavy (143M params)
SEED        = 42
NUM_CLASSES = 9

# ============================================================
# STEP 5: COPY BOTH TO LOCAL DISK
# ============================================================
LOCAL_AUG  = "/content/Augmented_dataset"
LOCAL_ORIG = "/content/Original_dataset"

if not os.path.exists(LOCAL_AUG):
    print("\nCopying Augmented dataset to local disk...")
    shutil.copytree(AUGMENTED_PATH, LOCAL_AUG)
    print("✅ Augmented dataset copied!")
else:
    print("✅ Augmented already on local disk!")

if not os.path.exists(LOCAL_ORIG):
    print("Copying Original dataset to local disk...")
    shutil.copytree(ORIGINAL_PATH, LOCAL_ORIG)
    print("✅ Original dataset copied!")
else:
    print("✅ Original already on local disk!")

# ============================================================
# STEP 6: LOAD DATASETS
# Train + Val → Augmented
# Test        → Original (100% unseen)
# ============================================================
print("\nLoading datasets...")

train_ds = tf.keras.utils.image_dataset_from_directory(
    LOCAL_AUG,
    validation_split=0.1,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds = tf.keras.utils.image_dataset_from_directory(
    LOCAL_AUG,
    validation_split=0.1,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

test_ds = tf.keras.utils.image_dataset_from_directory(
    LOCAL_ORIG,
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False       # Important — keep order for evaluation
)

class_names = train_ds.class_names
print(f"\nClasses ({len(class_names)}):", class_names)

train_batches = tf.data.experimental.cardinality(train_ds).numpy()
val_batches   = tf.data.experimental.cardinality(val_ds).numpy()
test_batches  = tf.data.experimental.cardinality(test_ds).numpy()

print(f"\nTrain batches : {train_batches} (~{train_batches * BATCH_SIZE} images)")
print(f"Val batches   : {val_batches}   (~{val_batches   * BATCH_SIZE} images)")
print(f"Test batches  : {test_batches}  (~{test_batches  * BATCH_SIZE} images)")

# ============================================================
# STEP 7: OPTIMIZATION — No cache to save RAM
# ============================================================
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.shuffle(500).prefetch(buffer_size=AUTOTUNE)
val_ds   = val_ds.prefetch(buffer_size=AUTOTUNE)
test_ds  = test_ds.prefetch(buffer_size=AUTOTUNE)

# ============================================================
# STEP 8: BUILD VGG19
# ============================================================
base_model = VGG19(
    weights='imagenet',
    include_top=False,
    input_shape=(224, 224, 3)
)
base_model.trainable = False  # Frozen — single phase

print(f"\nVGG19 loaded!")
print(f"Total layers : {len(base_model.layers)}")

# Build model
inputs = tf.keras.Input(shape=(224, 224, 3))

# VGG19 preprocessing — scales to [-1, 1]
x = tf.keras.applications.vgg19.preprocess_input(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(256, activation='relu')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)

model_vgg19 = models.Model(inputs, outputs)
model_vgg19.summary()

# Parameter summary
total_params     = model_vgg19.count_params()
trainable_params = sum([tf.size(w).numpy()
                        for w in model_vgg19.trainable_weights])
frozen_params    = total_params - trainable_params

print(f"\nTotal params    : {total_params:,}")
print(f"Trainable params: {trainable_params:,}  ← Your head")
print(f"Frozen params   : {frozen_params:,}  ← VGG19 base")
print(f"% Frozen        : {frozen_params/total_params*100:.1f}%")

# ============================================================
# STEP 9: COMPILE
# ============================================================
model_vgg19.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# ============================================================
# STEP 10: CALLBACKS
# ============================================================
checkpoint_path = '/content/drive/MyDrive/lemon/best_vgg19_augmented.keras'

callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=7,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        patience=3,
        factor=0.2,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        filepath=checkpoint_path,
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1
    )
]

# ============================================================
# STEP 11: TRAIN
# ============================================================
print("\n" + "="*50)
print("TRAINING VGG19...")
print("Train: Augmented (90%) | Val: Augmented (10%)")
print("Test : Original  (100% unseen)")
print("="*50)

history = model_vgg19.fit(
    train_ds,
    validation_data=val_ds,
    epochs=30,
    callbacks=callbacks
)

# ============================================================
# STEP 12: PLOT HISTORY
# ============================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(history.history['accuracy'],     'b-', label='Train Accuracy')
ax1.plot(history.history['val_accuracy'], 'r-', label='Val Accuracy')
ax1.set_title('VGG19 — Accuracy', fontweight='bold')
ax1.set_xlabel('Epoch'); ax1.set_ylabel('Accuracy')
ax1.legend(); ax1.grid(True, alpha=0.3)

ax2.plot(history.history['loss'],     'b-', label='Train Loss')
ax2.plot(history.history['val_loss'], 'r-', label='Val Loss')
ax2.set_title('VGG19 — Loss', fontweight='bold')
ax2.set_xlabel('Epoch'); ax2.set_ylabel('Loss')
ax2.legend(); ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/content/drive/MyDrive/lemon/vgg19_augmented_history.png', dpi=150)
plt.show()

# ============================================================
# STEP 13: EVALUATE ON ORIGINAL TEST SET
# ============================================================
print("\n" + "="*50)
print("EVALUATING ON ORIGINAL TEST SET (unseen)...")
print("="*50)

y_true, y_pred = [], []
for images, labels in test_ds:
    preds = model_vgg19.predict(images, verbose=0)
    y_pred.extend(np.argmax(preds, axis=1))
    y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

acc  = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average='macro', zero_division=0)
rec  = recall_score(y_true, y_pred, average='macro', zero_division=0)
f1   = f1_score(y_true, y_pred, average='macro', zero_division=0)

print(f"\n{'='*40}")
print(f"  VGG19 — Augmented Train / Original Test")
print(f"{'='*40}")
print(f"  Accuracy : {acc:.4f} ({acc*100:.2f}%)")
print(f"  Precision: {prec:.4f} ({prec*100:.2f}%)")
print(f"  Recall   : {rec:.4f} ({rec*100:.2f}%)")
print(f"  F1-Score : {f1:.4f} ({f1*100:.2f}%)")
print(f"{'='*40}")

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names))

# ============================================================
# STEP 14: CONFUSION MATRIX
# ============================================================
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds',
            xticklabels=class_names,
            yticklabels=class_names)
plt.title('VGG19 — Confusion Matrix (Augmented Train / Original Test)',
          fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/lemon/vgg19_augmented_cm.png', dpi=150)
plt.show()

# ============================================================
# STEP 15: SAVE MODEL & RESULTS
# ============================================================
model_vgg19.save('/content/drive/MyDrive/lemon/vgg19_augmented_final.keras')

results_text = f"""
VGG19 — Augmented Train / Original Test
========================================
Train : Augmented dataset (90% = ~8,100 images)
Val   : Augmented dataset (10% = ~900 images)
Test  : Original dataset  (100% = 1,354 images)
========================================
Accuracy : {acc:.4f} ({acc*100:.2f}%)
Precision: {prec:.4f} ({prec*100:.2f}%)
Recall   : {rec:.4f} ({rec*100:.2f}%)
F1-Score : {f1:.4f} ({f1*100:.2f}%)
========================================
"""

with open('/content/drive/MyDrive/lemon/vgg19_augmented_results.txt', 'w') as f:
    f.write(results_text)

print(results_text)

# ============================================================
# STEP 16: COMPARISON TABLE
# ============================================================
print("\n" + "="*60)
print("  EXPERIMENT 1 — Augmented Train → Original Test")
print("="*60)
print(f"  {'Model':<20} {'Accuracy':>10} {'F1-Score':>10}")
print("-"*60)
print(f"  {'EfficientNetB3':<20} {'99.93%':>10} {'?%':>10}")
print(f"  {'VGG19':<20} {acc*100:>9.2f}% {f1*100:>9.2f}%")
print("="*60)

# Clear memory
del model_vgg19
tf.keras.backend.clear_session()
gc.collect()
print("\n✅ Memory cleared!")
print("✅ All VGG19 outputs saved to Drive!")
```
